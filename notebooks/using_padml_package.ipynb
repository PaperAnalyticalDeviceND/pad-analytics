# 1. Introduction to `padanalytics` <a id="intro"></a>

**`padanalytics`** is a Python package that provides seamless integration with the PAD API for data fetching, management, and visualization. This notebook demonstrates how to use padanalytics to interact and explore data from the with the API.

## **Key Features:**
1. **Data Fetching**: Easily retrieve data from the PAD API using custom queries.
2. **Data Management**: Manage and manipulate data efficiently with built-in functions.
3. **Visualization**: Generate insightful visualizations directly from the data.


## **Type Functions:**

`padanalytics` has two main types of functions for handling data: **data retrieval functions** and **visualization functions**. Understanding the purpose and structure of these functions will help you use them more effectively in your analysis.

### **Data Retrieval Functions** (`get_<something>`)

- **Purpose**: These functions are designed to retrieve, process, and return data in the form of a `pandas` DataFrame.
- **Naming Convention**: Each function is named following the pattern `get_<description>`. The `<description>` portion describes the specific data or process being handled.
- **Output**: The returned data is always structured as a DataFrame, which can then be used for further analysis or as input for visualization functions.
  

### **Visualization Functions** (`show_<something>`)

- **Purpose**: These functions are intended to create visual representations of data using various plotting libraries like `matplotlib` or `seaborn`.
- **Naming Convention**: Each function is named following the pattern `show_<description>`, where `<description>` indicates the type of visualization or specific data insights being displayed.
- **Input**: These functions often take a DataFrame as input, allowing them to visualize the information contained in the retrieved data.
- **Output**: The output is typically a plot or graph that is rendered within the notebook.
